### idiomatic python
  - transformation code into beautiful, idiomatic python
  - replaces traditional index manipulation with python code looping idioms

```python
# looping over a range of numbers
for i in [0,1,2,3,4,5]:
    print(i ** 2,end=' ') 
print()
#pythonic-way
for i in range(6):
    print(i ** 2,end=' ')
```

```python
# looping over the collection(list)
li=[1,2,3,4,5]
for i in range(len(li)):
    print(li[i],end=' ')
print()
#pythonic-way
for i in li:
    print(i,end=' ')
```

```python
#looping from backwards
li = [1,2,3,4,5]
for i in range(len(li)-1,-1,-1):
    print(li[i],end=' ')
print()
#pythonic-way
for i in reversed(li):
    print(i,end=' ')
```

```python
# looping over the collection with index
li = [1,2,3,4,5]
for i in range(len(li)):
    print(i,'-->',li[i])
print()
#pythonic-way
for i,a in enumerate(li):
    print(i,'-->',a)
```

```python
# looping over two collections
li = [1,2,3,4,5]
a = [1,2,3,4,5,6,7,8]
n = min(len(li),len(a))
for i in range(n):
    print(li[i],' ',a[i])
print()
#pythonic-way
for i,a in zip(li,a):
    print(i,' ',a)
```

```python
# looping in sorted order
#pythonic-way
li = [8,1,25,16,9,2]
for i in sorted(li):
    print(i,end=' ')
print()
for i in sorted(li,reverse=True):
    print(i,end=' ')
```

### use lambda with filter

```python
def filterlist(li):
    a = []
    for i in li:
        if i % 2 == 0:
            a.append(i)
    return a
li = [1,2,3,4,5,6,7,8]
filterlist(li)
```

```python
li = [1,2,3,4,5,6,7,8]
lam_li = list(filter(lambda x : (x%2==0),li))
print(lam_li)
```

### lambda with map()
- map() function in python takes in a function and list as argument

```python
#functional approach
def squarelist(li):
    a=[]
    for i in li:
        a.append(i*2)
    return a
li = [1,2,3,4,5,6,7,8]
squarelist(li)
```

```python
#lambda with map
li = [1,2,3,4,5,6,7,8]
map_list = list(map(lambda x : x*2,li))
print(map_list)
```

### lambda with reduce()

```python
def sumlist(li):
    s=0
    for i in li:
        s += i
    return s
li=[1,2,3,4,5,6,7,8]
sumlist(li)
```

```python
from functools import reduce
li=[1,2,3,4,5,6,7,8]
s = reduce((lambda x,y:x+y),li)
print(s)
```
